## - Clean

```python
# Take copies before cleaning
archive_clean = archive.copy()
image_clean = image.copy()
tweet_clean = tweet.copy()
```

#### Quality:

#### `archive_clean`: has many columns values empty.

##### Define
Drop all columns have empty values, like (in_reply_to_status_id, in_reply_to_user_id, retweeted_status_id, retweeted_status_user_id, retweeted_status_timestamp).

##### Code

```python
archive_clean.drop(['in_reply_to_status_id','in_reply_to_user_id','retweeted_status_user_id', 'retweeted_status_id',\
                    'retweeted_status_timestamp'], axis='columns', inplace=True)
```

##### Test

```python
archive_clean.sample(5)
```

#### Tidiness:

#### First column in `tweet_clean` table contains many variables: we need three var (tweet_id, favorite_count, retweet_count)

##### Define
Extract the *id* and *favorite_count* and *retweet_count* variables from the first column using regular expressions and pandas' `str.extract` method. Drop the column when done.

##### Code

```python
tweet_clean['id_test'] = tweet_clean.tweet_json.str.extract(r'(, "id": \d+)', expand=True)
tweet_clean['retweet_test'] = tweet_clean.tweet_json.str.extract(r'(retweet_count": \d+)', expand=True)
tweet_clean['favorite_test'] = tweet_clean.tweet_json.str.extract(r'(favorite_count": \d+)', expand=True)
tweet_clean = tweet_clean.drop('tweet_json', axis=1)
```

```python
tweet_clean['tweet_id'] = tweet_clean.id_test.str.extract(r'(\d+)', expand=True)
tweet_clean['retweet_count'] = tweet_clean.retweet_test.str.extract(r'(\d+)', expand=True)
tweet_clean['favorite_count'] = tweet_clean.favorite_test.str.extract(r'(\d+)', expand=True)
tweet_clean = tweet_clean.drop(['id_test','retweet_test','favorite_test'], axis=1)
```

##### Test

```python
list(tweet_clean)
```

```python
tweet_clean.head(5)
```

#### Quality:

##### Define
Change `tweet_clean` columns data type to integer and float.

##### Code

```python
tweet_clean['tweet_id'] = tweet_clean.tweet_id.astype(float)
tweet_clean['retweet_count'] = tweet_clean.retweet_count.astype(int)
tweet_clean['favorite_count'] = tweet_clean.favorite_count.astype(int)
```

#### Tidiness:

##### Define
Condensing Dog Type columns.

##### Code

```python
dog_type = []

x = ['pupper', 'puppo', 'doggo', 'floof']
y = ['pupper', 'puppo', 'doggo', 'floof']

for row in archive_clean['text']:
    row = row.lower()
    for word in x:
        if word in str(row):
            dog_type.append(y[x.index(word)])
            break
    else:
        dog_type.append('None')

archive_clean['dog_type'] = dog_type
```
